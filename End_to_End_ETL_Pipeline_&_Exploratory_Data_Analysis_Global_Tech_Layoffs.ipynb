#Data Cleaning

```python
# ============================================
# CELL 7: CREATE CLEAN COPY
# ============================================
df_clean = df.copy()
print(f"Original shape: {df.shape}")
print(f"Clean copy shape: {df_clean.shape}")
print(f"Type: {type(df_clean)}")
print("Clean copy created ;)")
```

```python
# ============================================
# CELL 8: DROP DUPLICATES
# ============================================
before = len(df_clean)
df_clean = df_clean.drop_duplicates()
after = len(df_clean)
print(f"Rows before: {before}")
print(f"Rows after: {after}")
print(f"Duplicates removed: {before - after}")
```

```python
# ============================================
# CELL 9: DROP NULL CORE METRICS
# ============================================
before = len(df_clean)
df_clean = df_clean.dropna(
    subset=['total_laid_off', 'percentage_laid_off'],
    how='all'
)
after = len(df_clean)
print(f"Rows before: {before}")
print(f"Rows after: {after}")
print(f"Rows dropped: {before - after}")
```

```python
# ============================================
# CELL 10: FILL MINOR NULLS
# ============================================
text_cols = ['industry', 'country', 'stage']
for col in text_cols:
    df_clean[col] = df_clean[col].fillna('Unknown')
    print(f"'{col}' nulls filled ")

df_clean['funds_raised'] = df_clean['funds_raised'].fillna(0)
print("'funds_raised' nulls filled with 0 ")
```

```python
# ============================================
# CELL 11: CONVERT DATE TO DATETIME
# ============================================
print(f"Date type BEFORE: {df_clean['date'].dtype}")
df_clean['date'] = pd.to_datetime(
    df_clean['date'],
    errors='coerce'
)
print(f"Date type AFTER: {df_clean['date'].dtype}")
```

```python
# ============================================
# CELL 12: STRIP WHITESPACE
# ============================================
text_columns = df_clean.select_dtypes(include='object').columns
for col in text_columns:
    df_clean[col] = df_clean[col].str.strip()
    print(f"Whitespace stripped from '{col}' ")
```

```python
# ============================================
# CELL 13: FINAL AUDIT
# ============================================
print("=" * 50)
print("FINAL CLEANED DATASET SUMMARY:")
print("=" * 50)
print(f"\nFinal shape: {df_clean.shape}")
print(f"Original shape: {df.shape}")
print(f"Total rows removed: {df.shape[0] - df_clean.shape[0]}")

print("=" * 50)
print("REMAINING NULLS:")
print("=" * 50)
remaining_nulls = df_clean.isnull().sum()
remaining_nulls = remaining_nulls[remaining_nulls > 0]
if len(remaining_nulls) == 0:
    print("No nulls remaining ")
else:
    print(remaining_nulls)

print("=" * 50)
print("CLEANED DATA SAMPLE:")
print("=" * 50)
print(df_clean.head())
```

##Database Integration

```python
# ============================================
# CELL 14: INSTALL AND IMPORT DEPENDENCIES
# ============================================
import sqlite3
from sqlalchemy import create_engine, text
print("Libraries imported successfully :)")
```

```python
# ============================================
# CELL 15: CREATE DATABASE AND LOAD CLEAN DATA
# ============================================

engine = create_engine('sqlite:///layoffs_clean.db')

print("Database engine created ")
df_clean.to_sql(
    name='layoffs',
    con=engine,
    if_exists='replace',
    index=False
)

print(f"Data loaded into database successfully ")
print(f"Table 'layoffs' created with {len(df_clean)} rows")
```

```python
# ============================================
# CELL 16: VERIFY DATABASE LOAD
# ============================================

with engine.connect() as connection:

   Result = connection.execute(text("SELECT COUNT(*) FROM layoffs"))

   count = Result.fetchone()[0]

   print(f"Total rows in database: {count}")
   print(f"Total rows in df_clean: {len(df_clean)}")

   if count == len(df_clean):
       print("Row counts match — data loaded perfectly ;)")
   else:
       print("WARNING: Row count mismatch — investigate :(")
   preview = connection.execute(
        text("SELECT company, country, total_laid_off, date FROM layoffs LIMIT 5")
  )

print("\nFirst 5 rows from SQL database:")

print("-" * 50)
for row in preview:
        print(row)
print("\nDatabase verification complete ;)")
```

# ANALYTICAL SQL QUERYING

```python
print("Sprint 4: Analytical SQL Querying ;)")
print("Database: layoffs_clean.db")
print("Table: layoffs")
print("Rows available:", len(df_clean))
```

```python
# ============================================
# QUERY 1: TOP 20 COMPANIES BY TOTAL LAYOFFS
# ============================================

query_1 = """
SELECT
    company,
    country,
    SUM(total_laid_off)     AS total_layoffs,
    COUNT(*)                AS total_events,
    MAX(date)               AS most_recent_event
    FROM layoffs
WHERE total_laid_off IS NOT NULL
GROUP BY company, country
ORDER BY total_layoffs DESC
LIMIT 20;
"""

with engine.connect() as connection:
    query_1_result = pd.read_sql(text(query_1), connection)

print("=" * 55)
print("TOP 20 COMPANIES BY TOTAL LAYOFFS:")
print("=" * 55)
print(query_1_result.to_string(index=False))
```

```python
# ============================================
# QUERY 2: TOTAL LAYOFFS BY COUNTRY
# ============================================

query_2 = """
SELECT
    country,
    SUM(total_laid_off)           AS total_layoffs,
    COUNT(*)                      AS total_events,
    ROUND(AVG(total_laid_off), 0) AS avg_layoffs_per_event
FROM layoffs
WHERE total_laid_off IS NOT NULL
AND country != 'Unknown'
GROUP BY country
ORDER BY total_layoffs DESC
LIMIT 20;
"""

with engine.connect() as connection:
    query_2_result = pd.read_sql(text(query_2), connection)

print("=" * 55)
print("TOTAL LAYOFFS BY COUNTRY:")
print("=" * 55)
print(query_2_result.to_string(index=False))
```

```python
# ============================================
# QUERY 3: TOTAL LAYOFFS BY INDUSTRY
# ============================================

query_3 = """
SELECT
    industry,
    SUM(total_laid_off)           AS total_layoffs,
    COUNT(*)                      AS total_events,
    ROUND(AVG(total_laid_off), 0) AS avg_layoffs_per_event
FROM layoffs
WHERE total_laid_off IS NOT NULL
AND industry != 'Unknown'
GROUP BY industry
ORDER BY total_layoffs DESC;
"""

with engine.connect() as connection:
    query_3_result = pd.read_sql(text(query_3), connection)

print("=" * 55)
print("TOTAL LAYOFFS BY INDUSTRY:")
print("=" * 55)
print(query_3_result.to_string(index=False))
```

```python
# ============================================
# QUERY 4: YEARLY LAYOFF TREND
# ============================================

query_4 = """
SELECT
    strftime('%Y', date)          AS year,
    SUM(total_laid_off)           AS total_layoffs,
    COUNT(*)                      AS total_events
FROM layoffs
WHERE total_laid_off IS NOT NULL
AND date IS NOT NULL
GROUP BY strftime('%Y', date)
ORDER BY year ASC;
"""

with engine.connect() as connection:
    query_4_result = pd.read_sql(text(query_4), connection)

print("=" * 55)
print("YEARLY LAYOFF TREND:")
print("=" * 55)
print(query_4_result.to_string(index=False))
```

```python
# ============================================
# QUERY 5: ROLLING 3-MONTH LAYOFF TREND
# ============================================

query_5 = """
WITH monthly_totals AS (
    SELECT
        strftime('%Y-%m', date)       AS month,
        SUM(total_laid_off)           AS monthly_layoffs
    FROM layoffs
    WHERE total_laid_off IS NOT NULL
    AND date IS NOT NULL
    GROUP BY strftime('%Y-%m', date)
)

SELECT
    month,
    monthly_layoffs,
    SUM(monthly_layoffs) OVER (
        ORDER BY month
        ROWS BETWEEN 2 PRECEDING AND CURRENT ROW
    ) AS rolling_3_month_total
FROM monthly_totals
ORDER BY month ASC;
"""

with engine.connect() as connection:
    query_5_result = pd.read_sql(text(query_5), connection)

print("=" * 55)
print("ROLLING 3-MONTH LAYOFF TREND:")
print("=" * 55)
print(query_5_result.to_string(index=False))
```

```python
# ============================================
# CELL 25: CHART 1 — TOP 20 COMPANIES
# ============================================

fig_companies = px.bar(
    query_1_result,

    x='total_layoffs',
    y='company',


    orientation='h',

    color='total_layoffs',

    color_continuous_scale='Reds',


    title='Top 20 Tech Companies by Total Layoffs (2020-2026)',
    labels={
        'total_layoffs': 'Total Layoffs',
        'company': 'Company'
    },

    text='total_layoffs',

)

fig_companies.update_layout(
    height=600,

    xaxis_title='Total Layoffs',
    yaxis_title='Company',
    coloraxis_showscale=False,


    yaxis={'categoryorder': 'total ascending'},

    plot_bgcolor='white',

)

fig_companies.update_traces(
    texttemplate='%{text:,.0f}',

    textposition='outside'

)

fig_companies.show()
```
